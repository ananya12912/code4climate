```python
import pandas as pd

def userinput():
  year = input("enter the year: ")
  month = input("enter the month: ")
  date = input("enter the date: ")

  import os
  filelist = os.listdir('.')

  filename= 'clean_'+year+month+date+'.csv'

  testVar=0
  for files in filelist:
    if files == filename:
      testVar=1
  if testVar == 1:
    return filename
  else:
    print('you have chosen wrong!')
    print('try again!')
    return userinput()

data1 = pd.read_csv(userinput())

dataset = data1

PM25 = int(dataset["PM2.5"].mean())
print("The average PM2.5 concentration", PM25)


if PM25 > 0.0 and PM25 <= 12.0:
  print("This PM2.5 value is equal this AQI:", (50/12)*(PM25))
elif PM25 >= 12.1 and PM25 <=35.4:
  print("This PM2.5 value is equal this AQI:", ((100-51)/(35.4-12.1))*(PM25-12.1)+51)
elif PM25 >= 35.5 and PM25 <=55.4:
  print("This PM2.5 value is equal this AQI:", ((150-101)/(55.4-35.5))*(PM25-35.5)+101)
elif PM25 >= 55.5 and PM25 <=150.4:
  print("This PM2.5 value is equal this AQI:", ((200-151)/(150.4-55.5))*(PM25-55.5+151))
elif PM25 >= 150.5 and PM25 <=250.4:
  print("This PM2.5 value is equal this AQI:", ((300-201)/(250.4-150.5))*(PM25-150.5)+201)
elif PM25 >= 250.5 and PM25 <=350.4:
  print("This PM2.5 value is equal this AQI:", ((400-301)/(350.4-250.5))*(PM25-250.5)+301)
elif PM25 >= 350.5 and PM25 <=500.4:
  print("This PM2.5 value is equal this AQI:", ((500-401)/(500.4-350.5))*(PM25-350.5)+401)

PM10 = int(dataset["PM10"].mean())
print("The average PM10 concentration", PM10)

if PM10 > 0 and PM10 <= 54:
  print("This PM10 value is equal this AQI:", (50/54)*(PM10))
elif PM10 >= 55 and PM10 <=154:
  print("This PM10 value is equal this AQI:", ((100-51)/(154-55))*(PM10-55)+51)
elif PM10 >= 155 and PM10 <=254:
  print("This PM10 value is equal this AQI:", ((150-101)/(254-155))*(PM10-155)+101)
elif PM10 >= 255 and PM10 <354:
  print("This PM10 value is equal this AQI:", ((200-151)/(354-255))*(PM10-255)+151)
elif PM10 >= 355 and PM10 <=424:
  print("This PM10 value is equal this AQI:", ((300-201)/(424-355))*(PM10-355)+201)
elif PM10 >= 425and PM10 <=504:
  print("This PM10 value is equal this AQI:", ((400-301)/(504-425))*(PM10-425)+301)
elif PM10 >= 505 and PM10 <=604:
  print("This PM10 value is equal this AQI:", ((500-401)/(604-505))*(PM10-505)+401)


import matplotlib.pyplot as plt
x = dataset["hour"]
y = dataset['PM2.5']
z = dataset['PM10']
plt.plot(x, y,  x,z)
```
